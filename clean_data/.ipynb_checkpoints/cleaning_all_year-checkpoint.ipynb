```python
import pandas as pd
import numpy as np
```

```python
df_all = pd.DataFrame(columns=['Cat','Gender','Age','Year','Region','Pop'])
for year in [2013,2014,2015,2016,2017]:
    raw = pd.read_csv("../raw_data/East Central Data/Movie Purchaser Behavior EC {}.csv".format(year),skiprows=1)
    df = pd.DataFrame(raw)
    
    df = df.dropna().drop(df.columns[[1, 2, 3]], axis=1)
    # print(df.head(3))
    # replace ' ' in category names with "_"
    cat = list(i.strip().replace(' ','_').lower() for i in list(df.iloc[:,0]))
    # print("{} cat: {}".format(year,cat))

    # population for each category
    pop_list = []
    for index, row in df.iterrows():
        pop_list += [int(row[5])] + [int(row[8])]\
                    + [int(row[11])] + [int(row[14])]\
                    + [int(row[17])]+ [int(row[6])]\
                    +  [int(row[9])] + [int(row[12])]\
                    + [int(row[15])]+ [int(row[18])]

    # clean data table
    clean_df = pd.DataFrame(columns=['Cat','Gender','Age','Year','Region','Pop'])
    cat_list = []
    for i in cat:
        cat_list += [i]*10
    clean_df["Cat"] = cat_list
    clean_df['Gender'] = (['Male']*5+['Female']*5)*len(cat)
    clean_df['Age'] = ['28-24','25-34','35-44','45-54','55+']*2*len(cat)
    clean_df['Year'] = [year]*len(cat)*10
    clean_df['Region'] = ['Central East']*len(cat)*10
    clean_df['Pop'] = pop_list
    df_all = pd.concat([df_all,clean_df], ignore_index=True)
    print ("finish {}".format(year))
df_all
```

```python
df_all.to_csv('clean_all_year_EC_withCAT.csv', index = False)
```

```python
# convert category names to index
cat_list = list(df_all.Cat.unique())
cat_dict = dict(zip(cat_list,[str(i) for i in range(0,len(cat_list))]))
cat_dict
```

```python
for index, row in df_all.iterrows():
    row[0] = cat_dict[row[0]]
```

```python
df_all.to_csv('clean_all_year_EC_withIndex.csv', index = False)
```
